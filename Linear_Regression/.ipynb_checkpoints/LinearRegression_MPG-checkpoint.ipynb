```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.tools.plotting import scatter_matrix
import seaborn as sns
from sklearn import linear_model
```

```python
mpg_data.horsepower = mpg_data.horsepower.str.replace('?','0.0')
```

```python
mpg_data.horsepower = mpg_data.horsepower.astype(float)
```

```python
carname = []
mpg_car = mpg_data['carname']
for ind in range(len(mpg_car)):
    carname.append(mpg_car[ind].split()[0])
uniq_car = set(carname)
car_to_ix = { car:i for i,car in enumerate(uniq_car)}
mpg_data['carname'] = [car_to_ix[car.split()[0]] for car in mpg_car]
```
